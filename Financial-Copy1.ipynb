```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = r"C:\Users\Harsh\Downloads\Financial Analytics data.xlsx"
# Load the data from the specified sheet
df = pd.read_excel(file_path, sheet_name='Financial Analytics data')

# Display the first few rows and column names to understand the data
print(df.head())
print(df.columns)

# Extract relevant columns
mar_cap = df['Mar Cap - Crore']
sales_qtr = df['Sales Qtr - Crore']

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(mar_cap, sales_qtr, alpha=0.7)
plt.title('Market Capitalization vs. Quarterly Sales')
plt.xlabel('Market Capitalization (Crore)')
plt.ylabel('Quarterly Sales (Crore)')
plt.grid(True)
plt.show()
```
